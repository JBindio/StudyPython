```python
tips = sns.load_dataset('tips')
```

```python
# 박스 그래프
ax = plt.subplots(dpi=100)
ax = sns.boxplot(x='time',y='total_bill',data=tips)
ax.set_title('시간대별 매출비교 박스그래프')
ax.set_xlabel('시간대')
ax.set_ylabel('매출')
```
