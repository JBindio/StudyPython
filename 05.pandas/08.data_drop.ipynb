```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scientists = pd.read_csv('./data/scientists.csv')
ages = scientists['Age']
scientists
```

##### 데이터타입 변환

```python
born_datetime = pd.to_datetime(scientists['Born'], format='%Y-%m-%d')
print(born_datetime)
print()
died_datetime = pd.to_datetime(scientists['Died'], format='%Y-%m-%d')
print(died_datetime)
```

##### 데이터 추가

```python
scientists['Born_dt'], scientists['Died_dt'] = (born_datetime, died_datetime)
scientists
```

##### 행 삭제

```python
scientists.drop([0,3])
```

##### 열 삭제

```python
scientists.drop(['Born','Died'],axis=1)
```

##### 데이터 카피

```python
scientists2 = scientists.copy()
scientists2
```

##### Copy를 쓰지 않을경우 기본데이터에 영향

```python
list1 = [1,2,3]
list2 = list1
print(list2)

list2[0] = 4
print(list2)
print(list1)
list1 is list2
```
