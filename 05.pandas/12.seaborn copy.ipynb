```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
# 단변량 그래프
ax = plt.subplots(facecolor='w', dpi=100)
ax = sns.distplot(tips['total_bill']) # kde = False, hist = False
ax.set_title('Total Bill Histogram with Density Plot')
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
# 이변량 그래프
ax = plt.subplots(facecolor='w', dpi=100)
ax = sns.regplot(x = 'total_bill', y='tip', data = tips) # fit_reg = False 회귀선 제거
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 산점도 + 히스토그램 = jointplot
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y= 1.03)
```

```python
# 육각 그래프
joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y= 1.03)
```

```python
# 이차원 밀집도
ax = plt.subplots(dpi=100)
ax = sns.kdeplot(x='total_bill', y='tip', data=tips, shade=True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
# 박스 그래프
ax = plt.subplots(dpi=100)
ax = sns.boxplot(x='time',y='total_bill',data=tips)
ax.set_title('Boxplot of Total Bill by Time of Day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Total Bill')
```

```python
# 박스 그래프 변형(violinplot)
ax = plt.subplots(dpi=100)
ax = sns.violinplot(x='time',y='total_bill',data=tips)
ax.set_title('Boxplot of Total Bill by Time of Day')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Total Bill')
```

```python
fig = sns.pairplot(tips, height= 4, hue='sex')
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()
```

```python
# 색상추가 hue = '컬럼'
ax = plt.subplots(dpi=100)
ax = sns.violinplot(x='time',y='total_bill',hue='sex', 
                    data=tips, split=True)
```

```python
scatter = sns.lmplot(x='total_bill',y='tip',hue='sex', 
                     data=tips, fit_reg=False)
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, fit_reg=False,
           scatter_kws={'s':tips['size']*20}, height=5)
```

```python
sns.lmplot(x='total_bill',y='tip', data=tips, fit_reg=False,
           hue='sex', markers=['o','x'], height=5)
```
