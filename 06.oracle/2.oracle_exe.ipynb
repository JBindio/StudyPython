### 오라클 DB연동

```python
# conda install -c conda-forge cx_Oracle : 오라클 연동 모듈 설치
```

```python
import cx_Oracle as ora
import pandas as pd
```

```python
# 함수 만들기
def busan_06(sql):
    
    # 오라클 연결
    dsn = ora.makedsn('localhost', 1521, service_name = 'orcl')
    conn = ora.connect(user = 'busan_06', password = 'dbdb', dsn = dsn, encoding = 'utf-8')
    cur = conn.cursor()
    
    # SQL문
    sql = sql
    cur.execute(sql)
    
    # 행, 열 추출
    row = cur.fetchall()
    colname = cur.description
    
    # 연결 해제
    cur.close()
    conn.close()
    
    # 컬럼명 추출
    col = []

    for i in range(len(colname)):
        col.append(colname[i][0])
        
    # 데이터프레임 변환
    return pd.DataFrame(row, columns= col)
```

```python
busan_06('''
SELECT mem_id, mem_name, SUM(NVL(cart_qty,0)) "SUM_QTY"
  FROM member 
  LEFT OUTER JOIN
      (SELECT cart_member, cart_no, cart_qty
         FROM cart
        WHERE SUBSTR(cart_no,1,8) LIKE '200504%') S
    ON (mem_id = S.cart_member)
 GROUP BY mem_id, mem_name
 ORDER BY mem_id, mem_name 
 ''')
```
