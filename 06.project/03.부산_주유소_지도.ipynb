### 지역별 전기차 증가율에 따른 지역선정

```python
import pandas as pd
import folium
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import seaborn as sns
import platform
```

```python
# seaborn 설정 리셋
sns.reset_defaults()

# 폰트설정
if platform.system() == 'Windows' :
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
else :
    print('Check your OS System')
    
# 그래프에 마이너스 표시
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
# 정제한 최종 데이터 불러오기
busan_oil = pd.read_excel('./data/newdata/3. 주유소/부산_주유소_위경도_추출_최종.xlsx').drop(columns='Unnamed: 0')
busan_oil
```

```python
# 결측치 최종 확인
busan_oil.info()
```

```python
# 중복치 최종확인
busan_oil[busan_oil_map.duplicated()]
```

```python
# 중복치 제거후 중복치 최종확인
busan_oil = busan_oil.drop_duplicates()
busan_oil[busan_oil.duplicated()]
```

```python
# 주유소 마커 표시하기
busan_oil_map = folium.Map(
    # 초기 지도의 중앙점 좌표[위도, 경도]
    location=[35.150444, 129.059338],

    # 지도 스타일
    tiles='openstreetmap',
    # 초기 지도의 스케일
    zoom_start= 11
)

for i in range(len(busan_oil)):
    lat = busan_oil['위도'].iloc[i]
    lng = busan_oil['경도'].iloc[i]
    # print(lat,lng)
    # # general : gray, 3
    # # reserve : blue, 5
    # # generalDT : red, 5

    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        location= [lat, lng],
        # 마커 채우기 색상
        fill_color = 'darkorange',
        # 마커 채우기 불투명도
        fill_opacity = 1,
        fill=True,
        # 마커 테두리 색상
        color = 'black',
        # 마커 테두리 굵기
        weight =1,
        # 마커 지름
        radius= 3
    ).add_to(busan_oil_map)

busan_oil_map
```

```python
# 지도 추출
busan_oil_map.save('./data/부산_주유소.html')
```
