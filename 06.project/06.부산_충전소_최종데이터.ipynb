### 정제된 데이터 최종확인 및 합본 생성

```python
import pandas as pd
```

```python
# 정제된 데이터 불러오기
busan_charge_all = pd.read_csv('./data/newdata/2. 충전소/부산/부산_충전소_이용제한없음_최종.csv')
busan_charge_else = pd.read_csv('./data/newdata/2. 충전소/부산/부산_충전소_이용제한있음_최종.csv')
```

```python
# 데이터 결측치, 이상치 최종확인
busan_charge_all.info()
busan_charge_all.describe()
```

```python
# 데이터 중복치 최종확인
busan_charge_all[busan_charge_all.duplicated()]
```

```python
# 데이터 결측치, 이상치 최종확인
busan_charge_else.info()
busan_charge_else.describe()
```

```python
# 데이터 중복치 최종확인
busan_charge_else[busan_charge_else.duplicated()]
```

```python
# 최종확인 완료된 데이터 합치기
busan_charge = pd.concat([busan_charge_all,busan_charge_else]).reset_index(drop=True)
busan_charge
```

```python
# 정제 완료된 최종 데이터 저장
busan_charge_all.to_csv('./data/newdata/2. 충전소/부산_충전소_이용제한없음_최종.csv',index=False)
busan_charge_else.to_csv('./data/newdata/2. 충전소/부산_충전소_이용제한있음_최종.csv',index=False)
busan_charge.to_csv('./data/newdata/2. 충전소/부산_충전소_전체_최종.csv',index=False)
```
