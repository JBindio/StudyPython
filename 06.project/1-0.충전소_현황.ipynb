### 부산 충전소 현황

```python
import pandas as pd
import folium
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import seaborn as sns
import platform
```

```python
# seaborn 설정 리셋
sns.reset_defaults()

# 폰트설정
if platform.system() == 'Windows' :
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
else :
    print('Check your OS System')
    
# 그래프에 마이너스 표시
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
charge = pd.read_csv('./data/rawdata/전국_충전소_현황(기존데이터).csv')
charge
```

```python
# 지역 컬럼 이상 문자열 처리 
charge['지역'] = charge['지역'].str.strip()
```

```python
charge = charge.drop_duplicates()
```

```python
charge[charge.duplicated()]
```

```python
charge.info()
```

```python
# 필요지역(부산) 및 필요 컬럼 추출
charge_busan = charge[charge['지역'] == '부산광역시'].drop(columns=['급속충전량','비고','이용가능시간']).reset_index(drop=True)
```

```python
# 결측치, 이상치 확인
charge_busan.info()
charge_busan.describe()
```

```python
# 중복치 확인
charge_busan[charge_busan.duplicated()]
```

```python
# 충전기 ID값으로 보아 실제 중복치로 판단하여 중복치 제거
charge_busan.drop_duplicates(inplace=True)
charge_busan[charge_busan.duplicated()]
```

```python
# 이용자 제한의 결측치를 24시간 이용 가능으로 판단 하여 '제한없음'으로 값 재 정의
charge_busan['이용자 제한'].fillna('제한없음',inplace=True)
```

```python
# 시군구 컬럼의 결측치 확인
charge_busan[charge_busan['시군구'].isna()]
```

```python
# 정확한 결측치 확인을 위해 문자열 슬라이싱을 통한 재확인
charge_busan['주소2'] = charge_busan['주소'].str[10:]
charge_busan[charge_busan['주소2'] == '']
```

```python
# 결측치가 있는 행의 부실한 주소 변경을위한 인덱스 값 확인
charge_busan[charge_busan['주소2'] == ''].index
```

```python
# 인덱스 값 리스트화
busan_list = [923, 924, 948, 949, 950]
```

```python
# 탐색을 통한 주소 확인 및 리스트화 
busan_addr = ['부산시 동래구 낙민동 130-4',
              '부산광역시 동래구 복천동 381',
              '부산광역시 연제구 거제1동 419-2번지 거제역 3층 중앙',
              '부산광역시 연제구 거제동 48-5',
              '부산광역시 연제구 거제동 619-1']
```

```python
# 부실한 주소 처리
charge_busan.iloc[busan_list,-3] = busan_addr
```

```python
# 주소 처리 확인
charge_busan[charge_busan['주소2'] == '']
```

```python
# 시군구의 결측치 처리
charge_busan['시군구'] = charge_busan['주소'].str.split(' ', expand=True)[1]
```

```python
# 최종확인
charge_busan[charge_busan['주소2'] == '']
```

```python
# 확인차 사용한 주소2 컬럼 제거
charge_busan = charge_busan.drop(columns='주소2')
```

```python
# 이용자 제한이 있는 충전소와 없는 충전소 구분
charge_busan_all = charge_busan[charge_busan['이용자 제한'] == '제한없음'].reset_index(drop=True)
charge_busan_else = charge_busan[charge_busan['이용자 제한'] != '제한없음'].reset_index(drop=True)
```
