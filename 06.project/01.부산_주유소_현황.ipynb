```python
# csv 형식 주유소 전체 불러오기
df = pd.DataFrame()

read = glob.glob('./data/rawdata/주유소/*_주유소_현황.csv')
for i in read:
        data = pd.read_csv(i,encoding='euc-kr')
        df = pd.concat([df, data],ignore_index= True)
df
```

```python
# 주소 컬럼 불러온 후 결측치 제거
df_addr = df[['주소']].dropna()
df_addr
```

```python
# excel 형식 주유소 전체 불러오기
df2 = pd.DataFrame()

read = glob.glob('./data/rawdata/주유소/*_주유소_현황.xlsx')
for i in read:
        data = pd.read_excel(i)
        df2 = pd.concat([df2, data],ignore_index= True)
df2
```

```python
# 주소 컬럼 불러온 후 결측치 제거
df2_addr = df2[['주소']].dropna()
```

```python
# csv, excel 데이터 합치기
oil_addr = pd.concat([df_addr,df2_addr])
oil_addr
```

```python
# 합친 데이터 저장
oil_addr.to_csv('./data/rawdata/주유소/부산주유소_현황.csv',encoding='euc-kr',index=False)
```
