## 카카오API를 통한 부산 충전소 위치 위도,경도 추출

### 부산시내 이용제한이 없는 충전소 위치 위도,경도 추출

```python
import pandas as pd
import re
import requests, json

# 위경도 찾을 파일 불러오기
file_path = './data/2.충전소/충전소_부산_제한없음.csv'
busan_charge_all = pd.read_csv(file_path)

# 주소 전처리
pattern = r'\([^)]*\)'
text = []

for i in range(0, len(busan_charge_all['주소']),1):
    text.append(re.sub(pattern=pattern, repl='', string= busan_charge_all['주소'][i]))

text
busan_charge_all['주소'] = text

# 카카오API를 사용하여 주소 -> 좌표 변환
def get_location(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  # 'KaKaoAK '는 그대로 두시고 개인키만 지우고 입력해 주세요.
  # ex) KakaoAK 6af8d4826f0e56c54bc794fa8a294
  headers = {"Authorization": "KakaoAK 57da1f1bcfa0f4b06f0350e852287856"}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = {"위도": str(address['y']), "경도": str(address['x'])}
  address_name = address['address_name']

  return crd

# 파일에서 주소를 위경도로 변환해서 crd에 넣기
crd = []

for i in range(0, len(busan_charge_all['주소']), 1):
    try:
        crd.append(get_location(busan_charge_all['주소'][i]))
    except:
        crd.append({'위도':0,'경도':0})

# crd를 데이터프레임화 하여 원래 파일에 붙이기
busan_all = pd.DataFrame(crd)
busan_charge_all = pd.concat([busan_charge_all,busan_all], axis=1)
busan_charge_all
```

```python
# 위도,경도 출력되지 않은 주소 필터링
busan_charge_all[busan_charge_all['위도'] == 0]
```

```python
# 데이터 정제를 위한 내보내기
busan_charge_all_NaN.to_excel('./data/newdata/2. 충전소/부산_제한없음_위경도_추출_결측치.xlsx')
```
