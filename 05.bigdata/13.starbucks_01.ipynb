### 지도 라이브러리

```python
import folium
import json
```

```python
starbucks_map = folium.Map(
    # 초기 지도의 중앙점 좌표[위도, 경도]
    location=[37.573050, 126.979189],

    # 지도 스타일
    tiles='openstreetmap',
    # 초기 지도의 스케일
    zoom_start= 50
)
starbucks_map
```

```python
seoul_starbucks.columns
```

```python
for i in range(len(seoul_starbucks)):
    lat = seoul_starbucks['위도'].iloc[i]
    lng = seoul_starbucks['경도'].iloc[i]
    # print(lat,lng)
    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        location= [lat, lng],
        # 마커 채우기 색상
        fill_color ='green',
        # 마커 채우기 불투명도
        fill_opacity = 1,
        # 마커 테두리 색상
        color = 'yellow',
        # 마커 테두리 굵기
        weight =1,
        # 마커 지름
        radius= 3
    ).add_to(starbucks_map)

starbucks_map
```

```python
# 스타벅스 매장 타입별로 마커 표시하기
starbucks_map2 = folium.Map(
    # 초기 지도의 중앙점 좌표[위도, 경도]
    location=[37.573050, 126.979189],

    # 지도 스타일
    tiles='openstreetmap',
    # 초기 지도의 스케일
    zoom_start= 11
)

for i in range(len(seoul_starbucks)):
    lat = seoul_starbucks['위도'].iloc[i]
    lng = seoul_starbucks['경도'].iloc[i]
    # print(lat,lng)
    type = seoul_starbucks['매장타입'].iloc[i]
    # general : gray, 3
    # reserve : blue, 5
    # generalDT : red, 5
    color = 'black'
    if type == 'general':
        fillcolor = 'lightgreen'
        size = 3
    elif type == 'reserve':
        fillcolor = 'blue'
        size = 5
    else:
        fillcolor = 'red'
        size = 5
    
    # 지도에 표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        location= [lat, lng],
        # 마커 채우기 색상
        fill_color = fillcolor,
        # 마커 채우기 불투명도
        fill_opacity = 1,
        fill=True,
        # 마커 테두리 색상
        color = color,
        # 마커 테두리 굵기
        weight =1,
        # 마커 지름
        radius= size
    ).add_to(starbucks_map2)

starbucks_map2
```
