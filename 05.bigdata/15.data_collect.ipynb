```python
re = response.text
soup = BeautifulSoup(re, 'html.parser')
soup
```

```python
addr_list = soup.find_all('addr')
len(addr_list)
```

```python
telno_list = soup.find_all('telno')
len(telno_list)
```

```python
yadmnm_list = soup.find_all('yadmnm')
len(yadmnm_list)
```

```python
sidocdnm_list = soup.find_all('sidocdnm')
len(sidocdnm_list)
```
