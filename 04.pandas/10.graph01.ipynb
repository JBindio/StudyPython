##### 4분할 그래프 만들기

```python
dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
# 그래프 기본 틀 생성
fig = plt.figure(figsize = (7,5))
axes1 = fig.add_subplot(2, 2, 1) #row, col, index
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

# 그래프 데이터 입력
axes1.plot(dataset_1['x'], dataset_1['y'], 'r*')
axes2.plot(dataset_2['x'], dataset_2['y'], 'b*')
axes3.plot(dataset_3['x'], dataset_3['y'], 'y*')
axes4.plot(dataset_4['x'], dataset_4['y'], 'g*')
fig.tight_layout()
```

```python
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

```python
# 히스토그램(단일변량 그래프)
fig = plt.figure(figsize = (5,5), dpi = 100)
axes1 = fig.add_subplot(1, 1, 1)
axes1.hist(tips['total_bill'], bins=10, color='skyblue')
axes1.set_title('Histogram Of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

```python
# 산점도 그래프
scatter_plot = plt.figure(dpi=100)
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'], c='magenta')
axes1.set_title('Scatter Of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
plt.scatter(tips['total_bill'], tips['tip'], c='red')
```

```python
# 박스 그래프
boxplot = plt.figure(facecolor='w', dpi=100)
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot(
    [tips[tips['sex'] == 'Female']['tip'],
     tips[tips['sex'] == 'Male']['tip']],
     labels=['Female', 'Male'])

axes1.set_xlabel('Gender')
axes1.set_ylabel('Tip')
axes1.set_title('Boxplot Of Tips By Gender')
```

```python
# 다변량 산점도 그래프
# 함수 선언
def recode_gender(gender):
    if gender == 'Female':
        return 'red'
    else:
        return 'blue'

# apply 함수 사용
tips['color'] = tips['sex'].apply(recode_gender)
tips.head()
```

```python
scatter_plot = plt.figure(dpi=100)
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(x = tips['total_bill'],
              y = tips['tip'],
              s = tips['size'] * 10,
              c = tips['color'],
              alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Gender and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
# bar그래프 (카운트 비교)

bar_plot = plt.figure()
axes1 = bar_plot.add_subplot(1, 1, 1)

axes1.bar(tips['sex'], tips['total_bill'])
```
