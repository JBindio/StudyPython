##### 누락값 처리하기

```python
# 누락값 변경1 - 0으로 변경
ebola.fillna(0).iloc[:, 0:5].head()
```

```python
print(ebola['Cases_Guinea'].mean())
print(ebola['Cases_Guinea'].fillna(0).mean())
```

```python
# 누락값 변경2 - 이전 값으로 변경
print(ebola.iloc[0:5,0:5],'\n')
print(ebola.fillna(method='ffill').iloc[0:5,0:5])
```

```python
print(ebola['Cases_Guinea'].mean())
print(ebola['Cases_Guinea'].fillna(method='ffill').mean())
```

```python
# 누락값 변경3 - 이후 값으로 변경
print(ebola.iloc[0:5,0:5],'\n')
print(ebola.fillna(method='bfill').iloc[0:5,0:5])
```

```python
print(ebola['Cases_Guinea'].mean())
print(ebola['Cases_Guinea'].fillna(method='bfill').mean())
```

```python
# 누락값 변경4 - 이전,이후값의 평균값으로 변경
print(ebola.iloc[0:5,0:5],'\n')
print(ebola.interpolate().iloc[0:5,0:5])
```

```python
print(ebola['Cases_Guinea'].mean())
print(ebola['Cases_Guinea'].interpolate().mean())
```
